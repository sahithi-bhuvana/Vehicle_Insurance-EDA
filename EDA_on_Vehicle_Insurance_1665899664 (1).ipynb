### ii)Outliers

```python
#Generate a summary of count of all the outliers column wise
# this is the function to boxplot for easy visualisation of outliers..
def plot_boxplot(df,ft):
    df.boxplot(column=[ft])
    plt.grid(False)
    plt.show


print(plot_boxplot(customer_details_df,'age'))
```

```python
print(plot_boxplot(customer_details_df,'region_code'))

```

```python
# this functions finds outliers if present by IQR method..
def finding_outliers(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    iqr=q3-q1
    outlier=df[((df<(q1-1.5*iqr)) | (df>(q3+1.5*iqr)))]
    return outlier

outliers=finding_outliers(customer_details_df['age'])
print('Number of outliers :',len(outliers))
print('Maximum outlier value :',outliers.max())
print('Minimum outlier value :',outliers.min())

print("")
outliers=finding_outliers(customer_details_df['region_code'])
print('Number of outliers :',len(outliers))
print('Maximum outlier value :',outliers.max())
print('Minimum outlier value :',outliers.min())

print("")
print("from the above results we can conclude that customer details table doesn't have outliers..")
```

```python
##Replace all outlier values for numeric columns by mean. 
def replace_outlier(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    iqr=q3-q1
    upper=df[~(df>(q3+1.5*iqr))].max()
    lower=df[~(df<(q1-1.5*iqr))].min()
    df=np.where(df>upper,df.mean(),np.where(df<lower,df.mean(),df))
    return df

customer_policy_details_df['annual_premium(Rs)']=replace_outlier(customer_policy_details_df['annual_premium(Rs)'])
print(customer_policy_details_df['annual_premium(Rs)'])
```
